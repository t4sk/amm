```python
import math
import json
import numpy as np
import matplotlib.pyplot as plt

### Plot ###
N = 1000
# initial price
p0 = 1 
# volatility
v = 0.01
deltas = np.random.normal(0, v, N)

# Graph
plt.plot(deltas)
plt.show()

# Prepare data
MUL0 = 1 * 1e18
MUL1 = 1 * 1e18

data = []
total_0 = 0
total_1 = 0
for d in deltas:
    if d >= 0:
        data.append({
            "d_in": math.floor(d * MUL0),
            "zero_for_one": True
        })
        total_0 += d
    else:
        data.append({
            "d_in": -math.floor(d * MUL1),
            "zero_for_one": False
        })
        total_1 += -d

print(total_0)
print(total_1)

with open('../tmp/data.json', 'w') as f:
    json.dump(data, f)
```

```python
import math
import json
import numpy as np
import matplotlib.pyplot as plt

### Plot ###
N = 1000
# initial price
p0 = 1 
# volatility
v = 0.001
deltas = np.random.normal(0, v, N)

p = p0
ps = []
for dp in deltas:
    p += dp
    ps.append(p)

# Graph
plt.plot(ps)
plt.show()

# Prepare data
data = []

for p in ps:
    data.append({"p": math.floor(p * 1e18)})

with open('../tmp/data.json', 'w') as f:
    json.dump(data, f)
```
